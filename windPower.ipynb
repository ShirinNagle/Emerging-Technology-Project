#### Work related to Keras and training a model using the data set

```python
# Neural networks
import tensorflow.keras as kr

# Numerical arrays
import numpy as np

# Data frames
import pandas as pd

# Plotting
import matplotlib.pyplot as plt
```

```python
# plot style
plt.style.use("ggplot")

# plot size
plt.rcParams['figure.figsize'] = [14,8]
```

```python
# 1. Load data and separate, data to inout and output [1]
# want the input(speed) to map to output(power)
# create a training data frame with x(speed) and y(power) values
# input x values taken from csv file column speed, output y values taken from csv file column power
# don't know what the relationship is between x and y yet
# y = f(x)
train = pd.DataFrame()
train['x']= df["speed"]
train['y'] = df["power"]
#train['y'] = f(train['x'])
train
# Ensuring that correct values were in the correct variable.
#print(x) 
#print(y)
```

#### Model 1 - neural network with 1 neuron and activation = linear

```python
# 2. Define Keras model with linear activation
# Create a neural network with 1 neuron
model = kr.models.Sequential()
model.add(kr.layers.Dense(1, input_shape=(1,), activation = 'linear', kernel_initializer = 'ones', bias_initializer='zeros'))
model.compile('adam', loss='mean_squared_error')
```

```python
# Train the neural netowrk on our training data
model.fit(train['x'], train['y'], epochs=900)# epoch - an x value goes through 900 times
```

Tried an epoch value of 700 and an epoch value of 900, the loss for epoch 700  = 515, the loss for epoch 900  = 495,not much of a difference

```python
# take 4 handpicked values and see their predictions
# speed = 0.576 power should = 5.241
# speed = 0.701 power should = 9.746
# speed = 2.402 power should = 2.287
# speed = 5.305 power should = 11.664
model.predict([0.576,0.701 ,2.402,5.305])
```

The output above is quite inaccurate, for example a negative power output is incorrect. Trying with higher values and that seem to have a relationship of , if input = x, output = between x(4 ) and  x(5)

```python
# Trying with more values from higher up in the wind speed
#14.014=69.24
#19.695=97.173
#20.145=99.029
#22.497=109.932
#24.95 = 0
model.predict([14.014,19.695 ,20.145,22.497,24.95])
```

The above output is closer but not close enough because the model applied had a linear activation. Plot the predictions and actual results below, prediction is a linear progression.

```python
# Plot our predictions of the x values we trained on
plt.plot(train['x'], train['y'], label = 'actual')
plt.plot(train['x'], model.predict(train['x']), label = 'prediction')
plt.legend();
```

The plot of the actual data looks like a sigmoid shape  - the predicted data shape is linear. 
Trying sigmoid activation in the next model

#### Model 2 - neural network with 1 neuron and activation = sigmoid

```python
# 2. Define Keras model with linear activation
# Create a neural network with 1 neuron
model = kr.models.Sequential()
model.add(kr.layers.Dense(1, input_shape=(1,), activation = 'sigmoid', kernel_initializer = 'ones', bias_initializer='zeros'))
model.compile('adam', loss='mean_squared_error')
```

```python
# Train the neural netowrk on our training data
model.fit(train['x'], train['y'], epochs=900)# epoch - an x value goes through 900 times
```

The loss does not reduce by much using a sigmoid activation. All the predicted power values are very close to one

#### Model 3 - neural network with 2 layers activation = sigmoid, then linear

```python
model = kr.models.Sequential()
model.add(kr.layers.Dense(50, input_shape=(1,), activation = 'sigmoid', kernel_initializer = 'glorot_uniform', bias_initializer='glorot_uniform'))
model.add(kr.layers.Dense(1, activation = 'linear', kernel_initializer = 'glorot_uniform', bias_initializer='glorot_uniform'))
model.compile(kr.optimizers.Adam(lr=0.001), loss='mean_squared_error')
```

```python
# Train the neural netowrk on our training data
model.fit(train['x'], train['y'], epochs=900)# epoch - an x value goes through 900 times
```

Tried running a model.fit with 700 epochs and 900 epochs..only including 900 epoch result here. The loss difference between the two is minimal 700 =260 loss, 900 = 253.7 loss.

```python
# take 4 handpicked values and see their predictions - nowhere near the speed
# speed = 0.576 power should = 5.241
# speed = 0.701 power should = 9.746
# speed = 2.402 power should = 2.287
# speed = 5.305 power should = 11.664
model.predict([0.576,0.701 ,2.402,5.305])
```

```python
# Trying with more values from higher up in the wind speed
#14.014=69.24
#19.695=97.173
#20.145=99.029
#22.497=109.932
#24.95 = 0
model.predict([14.014,19.695 ,20.145,22.497,24.95])
```
